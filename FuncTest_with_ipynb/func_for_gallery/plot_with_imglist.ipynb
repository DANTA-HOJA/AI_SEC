Source: `{ch4_min_proj, outer_rect}_RGB_preprocess/20220825_AI002_palmskin_10dpf - Series005_fish_127_A_RGB`

```python
dir_path = Path("/home/rime97410000/ZebraFish_DB/{Data}_Processed/{20230424_Update}_Academia_Sinica_i505/{ch4_min_proj, outer_rect}_RGB_preprocess/20220825_AI002_palmskin_10dpf - Series005_fish_127_A_RGB")
path_list = list(dir_path.glob("**/*.tif"))

def get_image_name(img_path:Path):
    return str(img_path).split(os.sep)[-1]

path_list.sort(key=get_image_name)

row1_path_list = path_list[0:6]   # 01 ~ 06
row2_path_list = path_list[6:11]  # 07 ~ 11
row3_path_list = path_list[11:16] # 12 ~ 16
row4_path_list = path_list[16:]   # 17 ~ 22

def make_n_img(img_list, target_num):
    
    img_dict = { get_image_name(path) : cv2.imread(str(path)) for path in img_list }
    
    six_title_list = list(img_dict.keys())
    six_img_list = list(img_dict.values())
    
    while len(six_img_list) != target_num:
        six_img_list.append(np.ones_like(six_img_list[-1])*255)
        six_title_list.append("[Empty]")
    
    return six_title_list, six_img_list


fish_127_A_RGB_title_list = []
fish_127_A_RGB_img_list = []
for path_list in [row1_path_list, row2_path_list, row3_path_list, row4_path_list]:
    six_title_list, six_img_list = make_n_img(path_list, 6)
    fish_127_A_RGB_title_list.extend(six_title_list)
    fish_127_A_RGB_img_list.extend(six_img_list)
```

Function: `plot_with_imglist()`

```python
kwargs_plot_with_imglist = {
    "img_list"      : img_list,
    "row"           : 2,
    "column"        : 1,
    "fig_dpi"       : 100,
    "figtitle"      : "test",
    "subtitle_list" : [ f"subplot_{i}" for i in range(len(img_list)) ],
    # "font_style"    : "UbuntuMono-R.ttf",
    # "save_path"     : "out1.png",
    # "use_rgb"       : True,
    "show_fig"      : False,
    "verbose"       : True,
}
plt_show.plot_with_imglist(**kwargs_plot_with_imglist)
```

Function : `plot_with_imglist_auto_row()` ( auto row calculation )

```python
kwargs_plot_with_imglist_auto_row = {
    "img_list"      : fish_127_A_RGB_img_list,
    "column"        : 6,
    "fig_dpi"       : 200,
    "figtitle"      : f"RGB_preprocess, 20220825_AI002_palmskin_10dpf - Series005_fish_127_A_RGB",
    "subtitle_list" : fish_127_A_RGB_title_list,
    # "font_style"    : "UbuntuMono-R.ttf",
    "save_path"     : r"{RGB_preprocess}_fish_127_A_RGB.png",
    # "use_rgb"       : False,
    "show_fig"      : False,
    "verbose"       : True,
}
plt_show.plot_with_imglist_auto_row(**kwargs_plot_with_imglist_auto_row)
```
