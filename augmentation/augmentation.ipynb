## using ```secrets``` package

````python
import string
import secrets
from tqdm.auto import tqdm
alphabet = string.ascii_letters + string.digits

pw_set = set()

for i in tqdm(range(10**7), desc="gen pw"):
    password = ''.join(secrets.choice(alphabet) for i in range(8))
    if password in pw_set: tqdm.write(f"Repeated!!! : {password}")
    else: pw_set.add(password)
````

## hash table

````python
class HashTable:
    def __init__(self):
        self.capacity = 10  # 
        self.table = [[] for _ in range(self.capacity)]  # 

    def _hash(self, key):
        """"""
        return hash(key) % self.capacity

    def put(self, key, value):
        """"""
        hash_value = self._hash(key)
        for item in self.table[hash_value]: # encounter "Hash Collision"
            if item[0] == key:
                item[1] = value
                return
        self.table[hash_value].append([key, value])

    def get(self, key):
        """"""
        hash_value = self._hash(key)
        for item in self.table[hash_value]:
            if item[0] == key:
                return item[1]
        raise KeyError(key)
````

````python
import random

h = hash("123")
x = "0"
while (hash(x)%10) != (h%10):
    x = str(random.randint(0, 10**6))

print(f"x = '{x}', hash(x) = {hash(x)}, hash('123') = {hash('123')}")
````

# Main Function

````python
import os
import sys
import traceback
import re
import json
import string
import secrets
from typing import List, Dict, Tuple
from datetime import datetime
import argparse
from glob import glob
import pickle

from tqdm.auto import tqdm
from math import floor
import numpy as np
import pandas as pd
import cv2

import albumentations as A
from albumentations.augmentations.transforms import RandomBrightnessContrast, Sharpen
from albumentations.augmentations.blur.transforms import GaussianBlur

sys.path.append(r"C:\Users\confocal_microscope\Desktop\ZebraFish_AP_POS\modules") # add path to scan customized module
from logger import init_logger
````

````python
def get_fish_dataset_ID(path:str) -> Tuple[int, str]:
    # example_name : L_fish_9_A_selected_0.tiff
    fish_name = path.split(os.sep)[-1].split(".")[0]
    fish_name_list = re.split(" |_|-", fish_name)
    return int(fish_name_list[2]), fish_name_list[3], int(fish_name_list[-1])
````

````python
log = init_logger(r"Augmentation")
````

````python
transform = A.Compose(
    [
        # A.RandomCrop(width=240, height=320, p=0.5),
        A.Rotate(limit=40, p=0.5),
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.5),
        RandomBrightnessContrast(p=0.5),
        A.OneOf(
            [
                GaussianBlur(p=0.5),
                Sharpen(p=0.5),
            ], p=0.7)
    ]
)
````

````python
# Parameters
ap_dataset_root = r"C:\Users\confocal_microscope\Desktop\{Test}_DataSet"
dataset_name = r"{20230305_NEW_STRUCT}_Academia_Sinica_i409"
dataset_gen_method = "fish_dataset_horiz_cut_1l2_Mix_AP"
dataset_param_name = "DS_SURF3C_CRPS512_SF14_INT20_DRP100_RS2022"
class_to_apply = "S"
aug_times = 3


# Scan files
train_selected_dir = os.path.join(ap_dataset_root, dataset_name, dataset_gen_method, dataset_param_name, "train", "selected", class_to_apply)
dataset_imgorig_list = glob(os.path.normpath(f"{train_selected_dir}/*selected*.tiff"))
dataset_imgaug_list = glob(os.path.normpath(f"{train_selected_dir}/*aug*.tiff"))
## sorting
dataset_imgorig_list.sort(key=get_fish_dataset_ID)


# Informations of the dataset
summary = {}
summary["num_orig_img"] = len(dataset_imgorig_list)
summary["num_aug_exist"] = len(dataset_imgaug_list)
summary["num_total_before"] = summary["num_orig_img"] + summary["num_aug_exist"]
summary["aug_times"] = aug_times
summary["num_exp_img_gen"] = summary["num_orig_img"]*aug_times
summary["num_exp_total_after"] = summary["num_total_before"] + summary["num_exp_img_gen"]


# Augmentation uid_set
alphabet = string.ascii_letters + string.digits # possible characters in UID
uid_save_path = os.path.join(ap_dataset_root, dataset_name, "uid_frozenset.pkl")
try:
    with open(uid_save_path, 'rb') as f: # load the frozenset
        uid_frozenset = pickle.load(f)
        uid_set = set(uid_frozenset)
except:
    uid_set = set() # create a new set

# Create progress bar
pbar = tqdm(total=summary["num_exp_img_gen"], desc=f"Augmentation of class {class_to_apply} ")


# *** Start process ***
for path in dataset_imgorig_list:
    
    # example_name : L_fish_9_A_selected_0.tiff
    file_name = path.split(os.sep)[-1].split(".")[0] # extract file_name without extenasion
    fish_name_list = re.split(" |_|-", file_name)
    fish_name_with_class = "_".join(fish_name_list[:4]) # L_fish_9_A
    image = cv2.imread(path)
    
    for i in range(aug_times):
        transformed = transform(image=image)

        # Generating a UID 
        uid = ''.join(secrets.choice(alphabet) for i in range(8))
        while uid in uid_set: 
            uid = ''.join(secrets.choice(alphabet) for i in range(8))
        uid_set.add(uid)
        
        # write images
        wirte_name = f"{fish_name_with_class}_aug_{uid}.tiff"
        write_path = os.path.join(train_selected_dir, wirte_name)
        cv2.imwrite(write_path, transformed["image"])
        
        # update progressBar
        pbar.update(1)

pbar.close() # finished


# Update "summary"
summary["num_act_total_after"] = len(glob(os.path.normpath(f"{train_selected_dir}/*.tiff")))
if summary["num_act_total_after"] == summary["num_exp_total_after"]: summary["status"] = "passed"
else: raise ValueError("Total number of images does not match the expected value!")
print(json.dumps(summary, indent=4))


# Write "summary"
summary_save_path = os.path.join(ap_dataset_root, dataset_name, dataset_gen_method, dataset_param_name, "train", "selected")
time_stamp = datetime.now().strftime('%Y%m%d_%H_%M_%S')
log_path = os.path.join(summary_save_path, f"{{Logs_{class_to_apply}_{aug_times}}}_{{augmentation}}_{time_stamp}.log")
log_writer = open(log_path, mode="w")
log_writer.write(json.dumps(summary, indent=4))
log_writer.close()


# Write "uid_set"
uid_frozenset = frozenset(uid_set)
with open(uid_save_path, 'wb') as f:
    pickle.dump(uid_frozenset, f)
````
