```python
import os
import pandas as pd
from pathlib import Path
```

```python
path = Path(r"C:\Users\confocal_microscope\Desktop\WorkingDir\ZebraFish_DB\{Data}_Processed\{20230904_update}_Academia_Sinica_i831\data.xlsx")
df = pd.read_excel(path, index_col=[0])
df
```

```python
training_ratio = 0.8
train_ratio = 0.9
rand_seed = 2022
```

```python
training_df: pd.DataFrame = df.sample(frac=training_ratio, replace=False, random_state=rand_seed)
test_df: pd.DataFrame = df[~df.index.isin(training_df.index)]

train_df: pd.DataFrame = training_df.sample(frac=train_ratio, replace=False, random_state=rand_seed)
valid_df: pd.DataFrame = training_df[~training_df.index.isin(train_df.index)]
```

```python
print(len(df), len(training_df), len(train_df), len(valid_df), len(test_df))
```

```python
for idx in train_df.index:
    df.loc[idx, "dataset"] = "train"

for idx in valid_df.index:
    df.loc[idx, "dataset"] = "valid"

for idx in test_df.index:
    df.loc[idx, "dataset"] = "test"
```

```python
save_dir = Path(os.path.split(str(path))[0])

df.to_csv(save_dir.joinpath(f"dataset_{rand_seed}.csv"), encoding='utf_8_sig')
```
